```python
#@Author: Anirban Dutta
#@This code is for education purpose only, donot use for commercial use
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, accuracy_score, mean_absolute_error
from sklearn.linear_model import LinearRegression
import math
```

```python
data_for_college = pd.read_csv('predict_admission.csv')
```

```python
trim_college = [name.strip() for name in data_for_college.columns]
data_for_college.columns = trim_college
data_for_college.head()
```

```python
gre_score = data_for_college['GRE Score']/340*100
toefl_score = data_for_college['TOEFL Score']/120*100
cgpa_score = data_for_college['CGPA']/10*100
average_score = (gre_score+toefl_score+cgpa_score)
average_score_percent = average_score/3
average_score_percent
```

```python
data_for_college['Average Score'] = average_score_percent
```

```python
X = data_for_college[["Average Score"]]
y = data_for_college[["Chance of Admit"]]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20, random_state = 0)
```

```python
linear = LinearRegression()
linear.fit(X_train,y_train)
train_predictions = linear.predict(y_train)
train_predictions2 = linear.predict(X_train)
test_predictions = linear.predict(y_test)
test_predictions2 = linear.predict(X_test)
```

```python
print('Mean Absolute Error(MAE) of training:', mean_absolute_error(y_train, train_predictions2))
print('Mean Absolute Error(MAE) of testing:', mean_absolute_error(y_test, test_predictions2))
print('Mean Squared Error(MSE) of training:', mean_squared_error(y_train, train_predictions2))
print('Mean Squared Error(MSE) of testing:', mean_squared_error(y_test, test_predictions2))
print('Root Mean Squared Error(RMSE) of training:', np.sqrt(mean_squared_error(y_train, train_predictions2)))
print('Root Mean Squared Error(RMSE) of testing:', np.sqrt(mean_squared_error(y_test, test_predictions2)))
```

```python
plt.scatter(X_test, y_test, color = 'red',label="actual answer")  
plt.plot(X_test, test_predictions2, color = 'green',label='predicted answer')
plt.legend()
plt.title ('Machine Results')
plt.xlabel('Average Score')
plt.ylabel('Chance of Admit')
plt.show()
```
